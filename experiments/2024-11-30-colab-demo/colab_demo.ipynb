## 运行最小仿真
使用轻量配置快速跑通循环。

```python
from core.environment import SimulationConfig, LiftSimEnvironment

config = SimulationConfig(seed=123, floors=6, elevators=2, capacity=8, horizon_s=90)
env = LiftSimEnvironment(config)
obs = env.reset()
print('initial time', obs['time'])

# 固定选择 no-op 动作（留作占位，可改为策略动作列表）
for _ in range(20):
    actions = [0] * config.elevators
    obs, reward, done, info = env.step(actions)
    backlog = sum(obs['hall_call_up']) + sum(obs['hall_call_down'])
    print(f"t={obs['time']:3d} reward={reward:5.2f} backlog={backlog} action_mask={info['action_mask']}")
    if done:
        break
```
